**RESIZING IMAGES(USING PILLOW LIBRARY)**

```python
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')
img_resized = img.resize((200,200))
```

```python
img_resized.save('resized_dog.jpg')
```

```python
img_res = matimg.imread('/content/resized_dog.jpg')
img_plot = plt.imshow(img_res)
plt.show()
```

**CONVERTING RGB ->GREYSCALE (USING OPENCV)**

```python
#importing opencv libraries
import cv2
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
img.shape
```

```python
gray = cv2.cvtColor(img , cv2.COLOR_RGB2GRAY)
```

```python
gray.shape
```

Hence, no 3 colour layers

cv2.imshow() will display the image. But this will not be allowed in Google Colab.

```python
#lets save this
cv2.imwrite('dog_grayscale.jpg', gray)
```
